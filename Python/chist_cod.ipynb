"Глава 4 - Выбор понятных имен"

Основные правила
1) Имена должны быть содержательными.
2) Нельзя выбирать встроенные имена, которые испозуется в языке написания кода, 
к примеру: all, any, date, email, file, format, hash, id, input, list, min, max, 
object, open, random, set, str, sum, test и type и т.п. 
3) Не шутить с названием. Название должно быть правильно прочитано и понято любым программистом 
(вне зависимости от национальности и культуры)
4) Выбирать имена пригодыне для поиска (Ctrl+F)

Схемы регистра имен:
3 типа регистра;
1 - Змеиный регистр (snake_case) - разделяет слова символом подчеркивания. 
Все буквы записываются в нижнем регистре, а константы в верхнем "UPPER_SNAKE_CASE"
2 - Верблюжий регистр (camelCase) — слова записываются в нижнем регистре, 
но второе и следующие слова начинаются с заглавной.
3 - Схема Pascal (PascalCase) — каждое слово с заглавной, 
применяется в языке программирования Pascal (поэтому такое название). 
В проекте нужно использовать один регистр.


Соглашения об именах PEP 8:
Ссылка - https://www.python.org/dev/peps/pep-0008/#naming-conventions
1) Правила для соблюдения (необязательно) - 
2) Все буквы должны быть буквами ASCII — то есть латинскими буквами верхнего и нижнего 
регистров без диакритических знаков.
3) Имена модулей должны быть короткими и состоять только из букв нижнего регистра.
4) Имена классов необходимо записывать в схеме Pascal.
5) Имена констант следует записывать в верхнем змеином регистре.
6) Имена функций, методов и переменных записывают в нижнем змеином регистре.
7) Первый аргумент методов всегда должен называться self в нижнем регистре. 
Первый аргумент методов классов всегда должен называться cls в нижнем регистре.
8) Приватные атрибуты классов всегда начинают с символа подчеркивания ( _ ).
9) Публичные атрибуты классов никогда не начинают с символа подчеркива- ния (_).
Важнейший фактор удобочитаемости — не выбор схемы, а последовательность в применении этой схемы.


Длина имен:
Имена не должны быть слишком короткими или слишком длинными.
Слишком короткие имена - пример:
1) Одно- или двухбуквенное имя (например, g), вероятно, обозначает какое-то слово, начинающееся 
с этой буквы, но таких слов очень много.
2) Сокращенные имена вида mon — могут означать monitor, month, monster и множество других слов.
3) Имя из одного слова — например, start (начало) — может трактоваться по- разному: начало чего?
Слишком длинные имена:
Чем больше область видимости имени, тем более содержательным оно должно быть.
Если большой код (10.000стр+), то лучше расписать имя так, чтобы оно доносило суть.
Лучше лишние пояснения, чем их нехватка. 


Префиксы в именах:
Префиксы в именах иногда избыточны. Пример - Класс Cat с атри- бутом weight, 
очевидно, что weight (вес) относится к кошке (cat). 
Таким образом, имя catWeight будет слишком подробным и длинным.

Указывать единицы данных, чтобы избежать недорузумений.


Последовательные числовые суффиксы в именах:
1) Последовательные числовые суффиксы в именах указывают на то, что вам, возможно, 
стоит изменить тип данных переменной или включить дополнительную информацию в имя. 
Пример - payment1, payment2 и payment3 не сообщают чем они различаются, 
преобразовываем в переменную-кортеж - payments.
Если эти функции обладают разным поведением, оправдывающим определение отдельных функций, 
смысл чисел должен быть отражен в имени, например: makeLowPriorityPayment(amount) и makeHighPriorityPayment(amount) 
или make1stQuarterPayment(amount) и make2n dQuarterPayment(amount).


Выбирайте имена, пригодные для поиска:
Чтобы имя было найдено немедленно, создавайте уникальные имена с более длин-
ными именами переменных, которые содержат конкретную информацию.
Пример: Имя email слишком многозначно, поэтому лучше выбрать более содержательное имя: 
emailAddress, downloadEmailAttachment, emailMessage, replyToAddress и т. д


Избегайте шуток, каламбуров и культурных отсылок:
Культурные отсылки могут помешать ясно передать цель вашего кода.
Лучше всего писать код прямолинейно.


Не заменяйте встроенные имена:
Никогда не используйте встроенные имена Python для своих переменных. Напри- мер, присвоив переменной имя list или set, 
вы заместите функции Python list() и set(), что позднее может привести к появлению ошибок. Функция list() 
создает объекты списков, но ее замена может вызвать ошибку.

Часто заменяемые имена Python — all, any, date, email, file, format, hash, id, input, 
list, min, max, object, open, random, set, str, sum, test и type. Не берите их для своих идентификаторов.

Распространенная проблема — присваивание файлам .py имен, совпада- ющих с именами сторонних модулей.

Нужно помнить об опасности замены существующих имен в коде Python.


Худшие из возможных имен:
Если вам нужна переменная для хранения статистического отклонения температурных данных, 
используйте имя temperatureVariance. Выбор имени tempVarData будет неправильным.


Итог:
Имена не должны быть слишком короткими или слишком длинными. Лучше сделать имя избыточным, чем недостаточно содержательным.
Имя должно быть лаконичным, но информативным. Оно должно легко находиться функцией поиска Ctrl-F. 
То, насколько просто можно найти выбранное имя, свидетельствует о его уникальности.
Использование понятных имен — основополагающий фактор раз- работки качественного программного обеспечения.
